### Scatterplot

To plot a scatterplot of %Fat versus BMI, you'll need `two` arrays or lists: one for %Fat and another for BMI.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset: random (%Fat vs BMI)
np.random.seed(0) # for reproducibility
percent_fat = np.random.normal(25, 1.5, 100) # mean 25%, std 1.5%, 100 samples
BMI = np.random.normal(20, 3, 100)

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(percent_fat, BMI, alpha=0.75)

# Labels and title
plt.title('Scatter plot of % Fat vs BMI')
plt.xlabel('% Fat')
plt.ylabel('BMI')

# Show plot
plt.show()
```
