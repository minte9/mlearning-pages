### Scatterplot

When you have `two` continuous variables, you can graph them using a scatterplot.  
Scatterplots are great for displaying the `relationship` between two continuous variables.  
To plot a scatterplot of %Fat versus BMI, you'll need `two` arrays or lists: one for %Fat and another for BMI.  

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset: random (%Fat vs BMI)
np.random.seed(0) # for reproducibility
percent_fat = np.random.normal(25, 1.5, 100) # mean 25%, std 1.5%, 100 samples
BMI = np.random.normal(20, 3, 100)

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(percent_fat, BMI, alpha=0.75)

# Labels and title
plt.title('Scatter plot of % Fat vs BMI')
plt.xlabel('% Fat')
plt.ylabel('BMI')

# Show plot
plt.show()
```

### Bar Charts

Bar charts are a standard way to graph `discrete` variables.  
Each bar represents a distinct value, and the height represents its `proportion` in the entire sample.  

```python
import matplotlib.pyplot as plt

# Assume this is the result of a survey of 1000 households
# '0' for no car, '1' for one car, '2' for two cars, etc.
car_counts = [250, 400, 200, 100, 50]  # hypothetical distribution of car counts

# Corresponding number of cars in household
num_cars = ['0', '1', '2', '3', '4+']

# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(num_cars, car_counts, color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Count of Cars in Households')
plt.xlabel('Number of Cars')
plt.ylabel('Number of Households')

# Show plot
plt.show()
```

### Pie Charts

Pie charts are great for highlighting the `proportions` that groups comprise of the whole.  
You can use them `with` categorical, binary, and ordinal data that define groups in your sample.  

```python
import matplotlib.pyplot as plt

# Sample data: New car colors in 2012
colors = ['Silver', 'Black', 'White', 'Red', 'Blue', 'Other']
percentages = [23, 19, 16, 14, 12, 16]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=colors, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

# Add a title
plt.title('New Car Colors in 2012')

# Show the pie chart
plt.show()
```
